#### Importing dataset

```python
import pandas as pd
data = pd.read_csv(r"C:\Users\Marcus Padilha\Desktop\MACHINE_LEARNING\DATABASES\01_PRICE_OF_HOUSES_DATASET\kc_house_data.csv")
data.drop(['id','zipcode','lat','long'], axis =1, inplace =True)
data
```

#### Importing Libraries

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

#### Separating variables

```python
y = data['price']
x = data.drop(['price', 'date'], axis =1)
```

```python
x.dtypes
```

#### Selecting variables with the biggest square chi

```python
algorithm = SelectKBest(score_func = chi2, k = 2)
best_predictor_variables = algorithm.fit_transform(x, y)
```

#### Results

```python
print(f'Scores: {algorithm.scores_}')
print(f'Result of the transformation: {best_predictor_variables}')
```
