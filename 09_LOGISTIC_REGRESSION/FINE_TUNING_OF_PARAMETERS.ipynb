```python
import pandas as pd
data = pd.read_csv(r"C:\Users\Marcus Padilha\Desktop\MACHINE_LEARNING\DATABASES\03_DEODORANT_LIKING_DATA\Data_train_reduced.csv")
pd.set_option('display.max_columns',64)
pd.set_option('display.max_rows',64)
import warnings
warnings.filterwarnings('ignore')
data.drop('q8.20', axis = 1, inplace = True)
data.drop('q8.18', axis = 1, inplace = True)
data.drop('q8.17', axis = 1, inplace = True)
data.drop('q8.10', axis = 1, inplace = True)
data.drop('q8.9', axis = 1, inplace = True)
data.drop('q8.8', axis = 1, inplace = True)
data.drop('q8.2', axis = 1, inplace = True)
data.drop('Respondent.ID', axis = 1, inplace = True)
data.drop('Product', axis = 1, inplace = True)
data.drop('q1_1.personal.opinion.of.this.Deodorant', axis = 1, inplace = True)
data['q8.12'].fillna(data['q8.12'].median(), inplace = True)
data['q8.7'].fillna(data['q8.7'].median(), inplace = True)
data
```

```python
y = data['Instant.Liking']
x = data.drop('Instant.Liking', axis = 1)
```

```python
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
```

```python
# Defining the values that will be tested in LogisticRegression:
C_values = np.array([0.01, 0.1, 0.3, 0.5, 1, 2, 3, 5, 10, 20, 50, 100])
regularization = ['l1','l2']
grid_values = {'penalty': regularization, 'C' : C_values}
```

```python
# Creating the model:
model = LogisticRegression()
```

```python
# Creating the grids:
grid_logistic_regression = GridSearchCV (estimator = model, param_grid = grid_values, cv = 5)
```

```python
grid_logistic_regression.fit(x,y)
```

```python
# Print the best accuracy and the best parameters:
print (f'Best accuracy: {grid_logistic_regression.best_score_}')
print (f'C parameter: {grid_logistic_regression.best_estimator_.C}')
print (f'Regularization: {grid_logistic_regression.best_estimator_.penalty}')
```
