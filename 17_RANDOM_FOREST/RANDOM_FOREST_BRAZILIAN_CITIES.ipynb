## Verifying the best parameters

```python
import numpy as np
from sklearn.model_selection import RandomizedSearchCV
```

#### Defining the values that will be tested in the RandomForest:

```python
min_split = np.array([2,3,4])
max_level = np.array([3,5,7,9,11,14])
min_leaf = np.array([3,4,5,6])
grid_values = {'min_samples_split':min_split, 'max_depth':max_level, 'min_samples_leaf': min_leaf}
```

#### Creation of the model

```python
model = RandomForestRegressor(n_estimators = 100, n_jobs = -1)
```

#### Creating the grids

```python
gridRandomForest = RandomizedSearchCV(estimator = model, param_distributions = grid_values, cv = 3, n_iter = 50, n_jobs = -1)
gridRandomForest.fit(x_pca,y)
```
