# Verifying the best algorithm

### Importing the  commom libraries

```python
import numpy as np
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
```

### Setting the target and predictor variables

```python
y = data['StyleID']
x = data.drop('StyleID', axis = 1)
```

#### Seeting the train and test variables

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3)
```

## KNN Algorithm

#### Importing libraries

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import MinMaxScaler
```

#### Normalizing the predictor variables

```python
normalizer = MinMaxScaler(feature_range = (0, 1))
x_norm = normalizer.fit_transform(x)
```

#### Setting the parameter values will be tested

```python
K_values = np.array([3,5,7,9,11])
distance_calc = ['minkowski', 'chebyshev']
#P_values = np.array([1,2,3,4])
grid_values = {'n_neighbors': K_values, 'metric': distance_calc}#, 'p': P_values}
```

#### Creating the model

```python
model = KNeighborsClassifier()
```

#### Creating the grid

```python
gridKNN = GridSearchCV(estimator = model, param_grid = grid_values, cv = 3, n_jobs = -1)
```

#### Applying the grid

```python
gridKNN.fit(x_norm, y)
```

#### Result

```python
print('Best accuracy: ', gridKNN.best_score_)
print('Best K: ', gridKNN.best_estimator_.n_neighbors)
print('Best distance: ', gridKNN.best_estimator_.metric)
print('Best p value: ', gridKNN.best_estimator_.p)
```

## Naive Bayes

#### Importing libraries

```python
from sklearn.naive_bayes import GaussianNB
```

#### Creating the model

```python
model = GaussianNB()
```

#### Applying the model

```python
skfold = StratifiedKFold(n_splits = 10)
result = cross_val_score(model, x, y, cv = skfold)
```

#### Result

```python
print(f'Accuracy: {result.mean()}')
```

## Decision Tree

#### Importing libraries

```python
from sklearn.tree import DecisionTreeClassifier
```

#### Setting the parameter values will be tested

```python
min_splits = np.array([2,3,4,5,6,7,8])
max_level = np.array([3,4,5,6])
algorithm = ['gini', 'entropy']
grid_values = {'min_samples_split':min_splits, 'max_depth':max_level, 'criterion': algorithm}
```

#### Creating the model

```python
model = DecisionTreeClassifier()
```

#### Creating the grid

```python
gridDecisionTree = GridSearchCV(estimator = model, param_grid = grid_values, cv = 3, n_jobs = -1)
```

#### Applying the grid

```python
gridDecisionTree.fit(x,y)
```

#### Result

```python
print(f'Min Split: {gridDecisionTree.best_estimator_.min_samples_split}')
print(f'Max Depht: {gridDecisionTree.best_estimator_.max_depth}')
print(f'Selected Algorithm: {gridDecisionTree.best_estimator_.criterion}')
print(f'Accuracy: {gridDecisionTree.best_score_}')
```

## Using the best algorithm and its best parameters

#### Defining the algorithm and parameters

```python
model = DecisionTreeClassifier(min_samples_split = 2, max_depth = 4, criterion = 'entropy')
```

#### Aplying the model at the training data

```python
model.fit(x_train, y_train)
```

#### Verifying the accuracy in the test data

```python
result = model.score(x_test, y_test)
print(f'Accuracy: {result}')
```

#### Analysing the prediction

```python
prediction = model.predict(x_test)
prediction
```

#### Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
matrix =  confusion_matrix(y_test, prediction)
print(matrix)
```
