```python
import pandas as pd
data = pd.read_csv(r'C:\Users\marcu\Desktop\MACHINE_LEARNING\DATABASES\WISCONSIN_BREAST_CANCER\archive\data.csv')
pd.set_option('display.max_columns', 30)
data
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data['diagnosis'].replace(['B', 'M'], [0, 1], inplace = True)
```

```python
data['diagnosis']
```

```python
y = data['diagnosis']
x = data.drop(['diagnosis', 'id', 'Unnamed: 32'], axis = 1)
```

```python
x
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

# Normalizing the predictor variables
normalizer = MinMaxScaler(feature_range = (0,1))
x_norm = normalizer.fit_transform(x)

# splitting the training data and test data
x_training, x_test, y_training, y_test = train_test_split(x_norm, y, test_size = 0.3, random_state = 16)

# Creation of the model
model = KNeighborsClassifier (n_neighbors = 10)
model.fit(x_training, y_training)

# Score
result = model.score(x_test, y_test)

print(f'Accuracy: {result}')
```

```python
prediction = model.predict(x_test)
prediction
```

```python
from sklearn.metrics import confusion_matrix
matrix =  confusion_matrix(y_test, prediction)
print(matrix)
```

```python
predictionproba = model.predict_proba(x_test)
predictionproba
```

```python
probability = predictionproba[:, 1]
probability
```

```python
from sklearn.metrics import roc_curve
fpr, tpr, thresholds = roc_curve (y_test, probability)
```

```python
print(f'TPR: {tpr}')      
print(f'FPR: {fpr}')
print(f'Thresholds: {thresholds}')  
```

```python
import matplotlib.pyplot as plt
plt.scatter(fpr, tpr)
plt.show()
```

```python
from sklearn.metrics import roc_auc_score
print(roc_auc_score(y_test, probability))
```
