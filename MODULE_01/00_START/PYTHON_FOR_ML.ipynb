### Lambda Function

```python
def sumsquare(a,b):
    sumQ = a**2 + b**2
    return sumQ
```

```python
sumsquare(3,5)
```

```python
sumsquare02 = lambda a,b: a**2 + b**2
```

```python
sumsquare02(5,8)
```

### Funtion Map

```python
kph = [34,56,79,16,39,73,61]
```

```python
mph = []
for i in kph:
    mph.append(i/1.61)
mph
```

```python
mph02 = list(map(lambda x: x/1.61,kph))
```

```python
print(mph02)
```

```python
numbers = [10, 15, 21, 33, 42, 55]
```

```python
mapped_numbers = list((map(lambda x: x * 2 + 3, numbers)))
```

```python
print(mapped_numbers)
```
