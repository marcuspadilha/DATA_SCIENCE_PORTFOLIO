```python
y = data['Chance of Admit ']
x = data.drop(['Chance of Admit ','Serial No.'], axis =1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.3, random_state = 45)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet
linearModel = LinearRegression()
ridgeModel = Ridge()
lassoModel = Lasso()
elasticModel = ElasticNet()
linearModel.fit(x_train, y_train)
ridgeModel.fit(x_train, y_train)
lassoModel.fit(x_train, y_train)
elasticModel.fit(x_train, y_train)
resultLinear = linearModel.score (x_test, y_test)
resultRidge = ridgeModel.score (x_test, y_test)
resultLasso = lassoModel.score (x_test, y_test)
resultElastic = elasticModel.score (x_test, y_test)
resultGeral = {'Linear Score': resultLinear, 'Ridge Score':resultRidge, 'Lasso Score':resultLasso, 'ElasticNet Score':resultElastic}
for i in sorted(resultGeral, key = resultGeral.get, reverse = True):
    print(f'{i}: ', resultGeral[i])
    
```
