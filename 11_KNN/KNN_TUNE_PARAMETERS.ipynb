```python
import pandas as pd
data = pd.read_csv(r"C:\Users\Marcus Padilha\Desktop\MACHINE_LEARNING\DATABASES\04_WISCONSIN_BREAST_CANCER\data.csv")
pd.set_option('display.max_columns', 30)
data
```

```python
data['diagnosis'].replace(['B', 'M'], [0, 1], inplace = True)
```

```python
y = data['diagnosis']
x = data.drop(['diagnosis', 'id', 'Unnamed: 32'], axis = 1)
```

```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import GridSearchCV

# Normalizing the predictor variables
normalizer = MinMaxScaler(feature_range = (0,1))
x_norm = normalizer.fit_transform(x)

# Setting the values that will be tested
K_values = np.array([3,5,7,9,11])
distance_calc = ['minkowski', 'chebyshev']
P_values = np.array([1,2,3,4])
grid_values = {'n_neighbors':K_values, 'metric':distance_calc, 'p': P_values}

# Creation of the model
model = KNeighborsClassifier ()

# Creation of the grid:
gridKNN = GridSearchCV(estimator = model, param_grid = grid_values, cv = 5)
gridKNN.fit(x_norm, y)
```

```python
print('Best accuracy: ', gridKNN.best_score_)
print('Best K: ', gridKNN.best_estimator_.n_neighbors)
print('Best distance: ', gridKNN.best_estimator_.metric)
print('Best p value: ', gridKNN.best_estimator_.p)
```
