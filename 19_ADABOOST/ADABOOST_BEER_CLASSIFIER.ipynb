```python
import pandas as pd
data = pd.read_csv(r"C:\Users\Marcus Padilha\Desktop\MACHINE_LEARNING\DATABASES\09_BEER_RECIPES_DATASET\recipeData.csv", encoding='latin1')
data
```

```python
data.drop(['BeerID', 'URL','Name', 'Style','PrimingMethod','PrimingAmount', 'UserId' ], inplace =True, axis = 1)
data
```

```python
data['BoilGravity'].fillna(data['BoilGravity'].median(), inplace = True)
data['MashThickness'].fillna(data['MashThickness'].median(), inplace = True)
data['PitchRate'].fillna(data['PitchRate'].mean(), inplace = True)
data['PrimaryTemp'].fillna(data['PrimaryTemp'].median(), inplace = True)
data
```

```python
SugarScale_encode = pd.get_dummies(data['SugarScale'])
SugarScale_encode
BrewMethod_encode = pd.get_dummies(data['BrewMethod'])
BrewMethod_encode
```

```python
data = pd.concat([data, SugarScale_encode, BrewMethod_encode], axis = 1)
data
```

```python
data = data.drop(['SugarScale', 'BrewMethod'], axis =1)
data
```

```python
y = data['StyleID']
x = data.drop('StyleID', axis = 1)
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score
```

```python
model =  AdaBoostClassifier()
skfold = StratifiedKFold(n_splits = 3)
```

```python
result = cross_val_score(model, x, y, cv = skfold, n_jobs = -1)
print(result)
```
