```python
import pandas as pd
pd.set_option('display.max_columns', 23)
data = pd.read_csv(r"C:\Users\Marcus Padilha\Desktop\MACHINE_LEARNING\DATABASES\09_BEER_RECIPES_DATASET\recipeData.csv", encoding = "ISO-8859-1")
import warnings
warnings.filterwarnings('ignore')
data
```

```python
data02 = data.loc[data['StyleID'].isin([7, 10, 134, 9, 4, 30, 86, 12, 92, 6, 175, 39])]
data02
```

```python
data02.drop(['BeerID', 'URL','Name', 'Style','PrimingMethod','PrimingAmount', 'UserId' ], inplace =True, axis = 1)
data02
```

```python
data02['SugarScale'] = data02['SugarScale'].replace('Specific Gravity', 0)
data02['SugarScale'] = data02['SugarScale'].replace('Plato', 1)
```

```python
brewmethod_encode = pd.get_dummies(data02['BrewMethod'])
```

```python
data02.drop('BrewMethod', axis = 1, inplace = True)
```

```python
data03 = pd.concat([data02, brewmethod_encode], axis = 1)
```

```python
data03
```

```python
data = data03
```

```python
missing_percent = (data.isnull().sum()/len(data['StyleID']))*100
missing_percent
```

```python
data['BoilGravity'].fillna(data['BoilGravity'].median(), inplace = True)
```

```python
x_train = data[data['PitchRate'].notnull()]
x_train.drop('PitchRate', axis =1, inplace = True)

y_train = data[data['PitchRate'].notnull()]
y_train = y_train['PitchRate']

x_lack = data[data['PitchRate'].isnull()]
x_lack.drop('PitchRate', axis = 1, inplace = True)

y_lack = data[data['PitchRate'].notnull()]
y_lack = y_lack['PitchRate']
```

```python
x_train.drop('MashThickness', axis = 1, inplace = True)
x_train.drop('PrimaryTemp', axis = 1, inplace = True)
x_lack.drop('MashThickness', axis = 1, inplace = True)
x_lack.drop('PrimaryTemp', axis = 1, inplace = True)
```

```python
from sklearn.tree import DecisionTreeRegressor
# Aplying the model
model = DecisionTreeRegressor()
model.fit(x_train, y_train)
# Prediction of the new values
y_lack = model.predict(x_lack)
```

```python
data.PitchRate[data.PitchRate.isnull()] = y_lack
```

```python
missing = data.isnull().sum()
percent_missing = (data.isnull().sum() / len(data['StyleID'])) * 100
print(percent_missing)
```
