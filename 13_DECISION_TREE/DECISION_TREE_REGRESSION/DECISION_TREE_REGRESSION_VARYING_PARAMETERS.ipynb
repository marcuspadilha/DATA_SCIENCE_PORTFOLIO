```python
import pandas as pd
data =  pd.read_csv(r"C:\Users\Marcus Padilha\Desktop\MACHINE_LEARNING\DATABASES\02_COLLEGE_ADMISSION_DATASET\Admission_Predict.csv")
data
```

```python
data.drop('Serial No.', axis = 1, inplace = True)
```

```python
y = data['Chance of Admit ']
x = data.drop(['Chance of Admit '], axis = 1)
```

```python
import numpy as np
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV
```

```python
min_split = np.array([2,3,4,5,6,7])
max_level = np.array([3,4,5,6,7,9,11])
algorithm = ['mse', 'friedman_mse', 'mae']
grid_values = {'min_samples_split':min_split, 'max_depth': max_level, 'criterion': algorithm}
```

```python
model = DecisionTreeRegressor()
```

```python
gridDecisionTree = GridSearchCV (estimator = model, param_grid = grid_values, cv = 5)
gridDecisionTree.fit(x,y)
```

```python
print(f'Min split: {gridDecisionTree.best_estimator_.min_samples_split}')
print(f'Max depth: {gridDecisionTree.best_estimator_.max_depth}')
print(f'Selectes Algorithm: {gridDecisionTree.best_estimator_.criterion}')
print(f'R2 Coefficient: {gridDecisionTree.best_score_}')
```
